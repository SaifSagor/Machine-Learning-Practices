```python
#Making The Imports
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams['figure.figsize'] = (12.0 , 9.0)
```

```python
#pre processing input data
data = pd.read_csv('data.csv')

print(data)
```

```python
X = data.iloc[: , 0]
Y = data.iloc[: , 1]
```

```python
print(X)
```

```python
print(Y)
```

```python
#Plotting 
plt.scatter(X,Y)
plt.xlabel('Pizza Size')
plt.ylabel('Price')
plt.title('Pizza Price Prediction')
```

```python

#Building The model 
m = 0
c = 0

l = 0.0001     # Learning rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X))

# Performing Gradient Descent 
for i in range(epochs):
    Y_pred = m * X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y-Y_pred))  # Cost funtion derivative wrt m
    D_c = (-2/n) * sum(Y-Y_pred) # Cost function derivative wrt c

    m = m - l * D_m  #Update m
    c = c - l * D_c  #Update c

print(m, c)

```

```python
# Making Predictions
Y_pred = m * X + c

#Plotting the fitted line
plt.scatter(X,Y)
plt.xlabel('Pizza Size')
plt.ylabel('Price')
plt.title('Pizza Price Prediction')
plt.plot([min(X) , max(X)] , [min(Y_pred) , max(Y_pred)] , color = 'orange')

```
